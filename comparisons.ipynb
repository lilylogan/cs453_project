Currently, the file doesn't conduct ay real comparisons of our data but has info and examples of how they can be. 

Mathematical comparison techniques: 

    - Precision, Recall, and F1-score:
        Compare how many recommended songs are actually relevant by checking overlap with user preferences or known playlists.

    - Cosine Similarity Distribution:
        When both methods output ranked recommendations, compare their cosine similarity scores to the original song (for the Decision Tree method) or user profile (for the SVD method).

        Compute average similarity scores and plot histograms of similarity distributions.

    - Mean Average Precision (MAP) & Mean Reciprocal Rank (MRR):

        Use if there's an ideal ranked list to compare against.
        MAP measures how well relevant recommendations are positioned in the ranked list.
        MRR measures how soon a relevant recommendation appears in the ranked list.

Visualization methods:

```python
# Example of scatterplot of similarity scores:

# Plot the cosine similarity scores for recommendations from both methods for the same song

import matplotlib.pyplot as plt

songs = [1, 2, 3, 4, 5]  # Song indices
decision_tree_similarities = [0.89, 0.76, 0.72, 0.68, 0.60]
svd_similarities = [0.91, 0.79, 0.74, 0.70, 0.65]

plt.scatter(songs, decision_tree_similarities, color='blue', label='Decision Tree')
plt.scatter(songs, svd_similarities, color='red', label='SVD')
plt.xlabel("Recommended Song Index")
plt.ylabel("Cosine Similarity Score")
plt.legend()
plt.show()
```

```python
# Overlap bar chart: 

# Compare how many songs are recommended by both methods
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

methods = ["Decision Tree", "SVD"]
overlap_count = [3, 3]  # Number of overlapping songs
unique_count = [2, 2]  # Songs unique to each method

df_overlap = pd.DataFrame({"Method": methods * 2, "Count": overlap_count + unique_count, 
                           "Type": ["Overlap"] * 2 + ["Unique"] * 2})

sns.barplot(x="Method", y="Count", hue="Type", data=df_overlap)
plt.title("Overlap and Unique Songs in Recommendations")
plt.show()
```

```python
# Heatmap of recommendation similarities:

# Construct a heatmap of cosine similarities for different recommended songs

import seaborn as sns
import numpy as np

similarities_matrix = np.array([
    [1.0, 0.89, 0.76, 0.72, 0.68],  # Decision Tree
    [1.0, 0.91, 0.79, 0.74, 0.70]   # SVD
])

sns.heatmap(similarities_matrix, annot=True, cmap="coolwarm", xticklabels=["Song1", "Song2", "Song3", "Song4", "Song5"],
            yticklabels=["Decision Tree", "SVD"])
plt.title("Similarity Scores Heatmap")
plt.show()
```
